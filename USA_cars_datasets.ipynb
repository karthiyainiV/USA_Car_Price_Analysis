Outliers finding:

```python
# Calculate 25th percentail of price #
percentaile25 = Car_data_set['price'].quantile(0.25)
# Calculate 75th percentail of price #
percentaile75 = Car_data_set['price'].quantile(0.75)
# Calculate 1qr range #
iqr = percentaile75 - percentaile25
# calculate upper and lower threshold of outlier #
upper = percentaile75 + 1.5*iqr  ; print("Upper Limit:",upper)
lower = percentaile25 - 1.5*iqr ; print("Lower  Limit:", lower )
```

```python
# Isolate outliers on lower end #
Car_data_set[Car_data_set['price'] <= lower]
```

```python
# Isolate outliers on upper end #
print(" ")
Car_data_set[Car_data_set['price'] > upper]
```

```python
import matplotlib.pyplot as plt
import numpy as np
import statsmodels as stats



colors = np.where(Car_data_set['price'] <= lower, 'r', 'b')

fig, ax = plt.subplots(figsize=(16, 8))
ax.scatter(y = Car_data_set['brand'], x = Car_data_set['price'], c=colors)
ax.set_xlabel('Brand')
ax.set_ylabel('Price Range')
ax.set_title("Price Range as per Brand")
plt.show()

```

```python
# Remove outliers #
Car_data_set_new = Car_data_set[Car_data_set['price'] > upper]

#Recalculate the mean and median values #
Mean_new = np.mean(Car_data_set_new['price']) ; print(Mean_new)

Median_new = np.median(Car_data_set_new['price']) ; print(Median_new)
```
